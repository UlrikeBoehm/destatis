# Physics Students in Germany (1997-2019)

Die Daten kommen von [statistischen Bundesamt](https://www-genesis.destatis.de/genesis/online). 

## Students: Germany, semester, nationality, sex, subject

### Data Loading

Daten werden vom `21311-0003,csv` File eingelesen und in der Liste `array` gespeichert.

```python
import csv
with open('Data/21311-0003.csv', newline='') as f:
    reader = csv.reader(f, delimiter=';')
    array = []
    for row in reader:
        array.append(row)
        print(row)
```

### Data Wrangling and Cleaning

Nach der erste Inspektion der Daten wird klar, dass die gewuenschte Datensaetze in den Zeilen 1, 10, 11 und 12 von Reihe 8 bis 28 liegen. Die einzenen Eintraege muessen zudem noch von `str` (string) Format in den `int` (integer) Format umgewandelt werden.  

```python
array = array[7:28] # loeschen von Kopf- und Fusszeilen
len(array)
```

```python
x = np.arange(21)+1 # we do not want to start with zero
x
```

```python
female = []
for i in range(0, len(array)):
    b = array[i]
    b[10] = int(b[10])
    female.append(b[10])
female;
```

```python
len(female)
```

```python
male = []
for i in range(0, len(array)):
    b = array[i]
    b[9] = int(b[9])
    male.append(b[9])
male;
```

```python
total = []
for i in range(0, len(array)):
    b = array[i]
    b[11] = int(b[11])
    total.append(b[11])
total;
```

### Data Plotting and Visualization

```python
plt.plot(x,female,'-or', label='Frauenanteil')
plt.title('Frauenanteil der letzten Jahre')
plt.xlabel('Years')
plt.ylabel('Frauenanteil')

plt.legend();
```

```python
plt.plot(x,male,'-ob', label='Frauenanteil')
plt.title('Maenneranteil der letzten Jahre')
plt.xlabel('Years')
plt.ylabel('Maenneranteil')

plt.legend();
```

```python
plt.plot(x,total,'-og', label='Gesamtanzahl')
plt.title('Gesamtanteil der letzten Jahre')
plt.xlabel('Years')
plt.ylabel('Gesamtanteil')

plt.legend();
```

```python
plt.plot(x,female,'-or', label='Frauenanteil')
plt.plot(x,male,'-ob', label='Maenneranteil')
plt.plot(x,total,'-og', label='Gesamtanzahl')
plt.title('Personenanteil der letzten Jahre')
plt.xlabel('Years')
plt.ylabel('Personenanteil')

plt.legend();
```

#### Relative numbers in comparison

```python
femalep = []
for i in range(0, len(female)):
    femalep.append(female[i]/total[i]);
femalep;
```

```python
malep = []
for i in range(0, len(male)):
    malep.append(male[i]/total[i]);
malep;
```

```python
plt.plot(x,femalep,'-or', label='Frauenanteil')
plt.plot(x,malep,'-ob', label='Maenneranteil')
plt.title('Anteil innerhalb der letzten Jahre in Prozent')
plt.xlabel('Years')
plt.ylabel('Personenanteil')

plt.legend();
```
