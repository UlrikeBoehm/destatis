### Physikprofessorinnen 

#### Data Loading

Daten werden von einem CSV File eingelesen und in der Liste `array` gespeichert.

```python
import csv
with open('Data/21341-0003.csv', newline='') as f:
    reader = csv.reader(f, delimiter=';')
    array = []
    for row in reader:
        array.append(row)
        print(row)
```

#### Data Wrangling and Cleaning

Nach der erste Inspektion der Daten wird klar, dass die gewuenschte Datensaetze in Reihe 5-8 von Zeile 2 bis 24 liegen. Die einzenen Eintraege muessen zudem noch von `str` (string) Format in den `int` (integer) Format umgewandelt werden.  

```python
x = array[5] # Years
x = x[2:24]
for i in range(0, len(x)): 
    x[i] = int(x[i])
```

```python
y_male = array[6] # maennliche Naturwissenschaftler
y_male = y_male[2:24]
for i in range(0, len(y_male)): 
    y_male[i] = int(y_male[i])
```

```python
y_female = array[7] # weibliche Naturwissenschaftler
y_female = y_female[2:24]
for i in range(0, len(y_female)): 
    y_female[i] = int(y_female[i])
```

```python
y_total = array[8] # ingesamt
y_total = y_total[2:24]
for i in range(0, len(y_total)): 
    y_total[i] = int(y_total[i])
```

#### Data Plotting and Visualization

```python
plt.plot(x,y_female,'-og', label='Frauenanteil')
plt.title('Frauenanteil der letzten Jahre')
plt.xlabel('Years')
plt.ylabel('Frauenanteil')

plt.legend();
```

```python
plt.plot(x,y_male,'-ob', label='Herrenanteil')
plt.title('Maenneranteil der letzten Jahre')
plt.xlabel('Years')
plt.ylabel('Maenneranteil')

plt.legend();
```

```python
plt.plot(x,y_total,'-or', label='Gesamtanzahl')
plt.title('Gesamtanteil der letzten Jahre')
plt.xlabel('Years')
plt.ylabel('Gesamtanteil')

plt.legend();
```

```python
plt.plot(x,y_female,'-og', label='Frauenanteil')
plt.plot(x,y_male,'-ob', label='Herrenanteil')
plt.plot(x,y_total,'-or', label='Gesamtanzahl')
plt.title('Personenanteil der letzten Jahre')
plt.xlabel('Years')
plt.ylabel('Personenanteil')

plt.legend();
```
